# EDA on Features

Get a bar plot for the 'Sex' column using seaborn.

```
sns.countplot(x='Sex', data=df_train)
```

Use seaborn to build bar plots of the Titanic dataset feature 'Survived' split over the feature 'Sex'.

```
sns.factorplot(x='Survived', col='Sex', kind='count', data=df_train)
```

How could you get the same information with pandas?

```
df_train.groupby(['Sex']).Survived.sum()
```

```
sns.factorplot(x='Survived', col='Age', kind='count', data=df_train);
```

Use seaborn to build bar plots of the Titanic dataset feature 'Survived' split over the feature 'Embarked'.

```
sns.factorplot(x='Survived', col='Embarked', kind='count', data=df_train);
```

Use seaborn to plot a histogram of the 'Fare' column of df_train.

```
sns.distplot(df_train.Fare, kde=True, hist=False)
```

Use seaborn to plot a histogram of the 'Age' column of df_train. You might need to drop null values before doing so.

```
df_train_drop = df_train.dropna('Age') # What pandas functions drops null values?
sns.distplot(df_train_drop.Age)
```

What proportion of women survived?

What proportion of  men survived?

```
print(df_train[df_train.Sex == 'female'].Survived.sum()/df_train[df_train.Sex == 'female'].Survived.count())
print(df_train[df_train.Sex == 'male'].Survived.sum()/df_train[df_train.Sex == 'male'].Survived.count())
```

Use the DataFrame method .describe() to check out summary statistics of 'Fare' as a function of survival.

```
df_train.groupby('Survived').Fare.describe()
```

How did people that paid a higher wage do compared to those that paid the average wage?

Let's see if age was a factor in survival.

Use seaborn to plot a scatter plot of 'Age' against 'Fare', colored by 'Survived'.

```
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df_train)
```
